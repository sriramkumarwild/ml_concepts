```python
import numpy as np
from numpy.linalg import inv
%pylab inline
```

```python
def get_modelparams(X, y):
    # w = (X^T * X)^-1 * X^T * y
    w = np.dot(np.dot(inv(np.dot(X,X.T)),X),y)
    
    return w

def get_polyfeatures(x, degree):
    Data  = np.ones([len(x)])
    for i in range(0,int(degree)):
        Data  = np.append(Data, x**i)
        Data  = np.resize(Data, (i+1,len(x)))
    
    return Data
```
